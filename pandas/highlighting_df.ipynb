```python
import pandas as pd
```

```python
df = pd.read_csv('stocks2.csv').sample(frac = 0.5)
```

```python
# highlighting min, max and NaN values
df.style.highlight_min(color='green').highlight_max(color = 'magenta').highlight_null()
```

```python
#Color the background in a gradient style.
df.style.background_gradient()
```

```python
# Making table borders blue with text color as green.
df.style.set_properties(**{'border': '1.3px solid blue',
                          'color': 'green'})
```
