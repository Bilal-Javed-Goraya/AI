A Parallel coord plot is used to `analyze multivariate numerical data` <br> We can compare observations across multiple numerical variables

```python
import pandas as pd
```

```python
df = pd.read_csv('iris.csv')
df.head(2)
```

```python
pd.plotting.parallel_coordinates(
    df, class_column='Name', 
    color=('#556270', '#4ECDC4', '#C7F464'))
```

Observation: `SepalWidth` in case of `Iris-setosa` is greater than <br>`PetalWidth` while it's not the case for other 2
