```python
import pandas as pd
```

```python
df = pd.read_csv('weather_data.csv')
df.head(2)
```

```python
# Let's group the data based on city
for city, city_df in df.groupby('city'):
    print(city)
    print(city_df)
```

```python
# Now let's check the max temperature and windspeed in each city, you can also try groups.mean()
df.groupby('city').max() 
```

```python
# üí´ getting all the analytics in one go üí• 
df.groupby('city').describe()
```
