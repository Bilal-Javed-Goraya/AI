##### ‚ùáÔ∏è Pandas üêº: Hexagonal Bin plots
Hexbin plots can be a useful alternative to scatter plots if your data are <br> too dense to plot each point individually.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=["a", "b"])
df["b"] = df["b"] + np.arange(1000)
```

```python
df.plot.hexbin(x="a", y="b", gridsize=25)
```
