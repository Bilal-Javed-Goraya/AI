```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('llms.csv')

plt.figure(figsize=(7, 7))
x, y = data['year'], data['parameters']

plt.scatter(x, y, s=data['parameters']*40, c=data['colors'], 
            cmap="Accent", alpha=0.6, edgecolors="white", linewidth=2);
 
plt.yticks(ticks = [0.01, 0.1, 1, 10, 100, 1000])
plt.yscale('log')
plt.xlabel("Year"), plt.ylabel("# Parameters in billions")
plt.title("Evolution of LLMs")

# Annotate the bubbles
for i, label in enumerate(list(data['label'])):
    plt.annotate(label, (x[i], y[i]), ha='center', va='center')
```
