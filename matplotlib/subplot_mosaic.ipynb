## ðŸ”˜ Matplotlib: `Subplot Mosaic` 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Let's define quantities to be plotted
x = np.linspace(0, 10)
functions_of_x = {'A': x, 'B': np.sin(x), 'C': np.cos(x), 'D': np.tan(x)}

# plot titles and colors dictionaries
colors = {'A': 'red', 'B': 'green', 'C': 'blue', 'D': 'magenta'}
titles = {'A': 'x', 'B': 'sin(x)', 'C': 'cos(x)', 'D': 'tan(x)'}
```

```python
# Specifying how the subplot axes should be laid out on the figure
mosaic = """
    ABD
    CCD
    """
# ðŸ‘† You can play around by changing the layout 
```

```python
fig = plt.figure(figsize=(5, 5), constrained_layout=True)

# axis_dict contains the mosaic axes against the keys 'A', 'B', 'C', 'D'
# over which we draw our subplots
ax_dict = fig.subplot_mosaic(mosaic)

for label, axes in ax_dict.items():
    axes.plot(x, functions_of_x[label], color=colors[label])
    axes.set_title(titles[label])
```
