##### ❇️ Matplotlib: Polar scatter plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Compute areas and colors
N = 150
r = 2 * np.random.rand(N)
theta = 2 * np.pi * np.random.rand(N)

# area of each circle representing each data point
area = 200 * r**2 

colors = theta

fig = plt.figure()

# Setting up projection as 'polar', So that
# scatter plot is created on polar coordinates
ax = fig.add_subplot(projection='polar')

c = ax.scatter(theta, r, c=colors, s=area, cmap='hsv', alpha=0.75)
```
