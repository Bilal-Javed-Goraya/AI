##### ❇️ Horizontal vs Vertical Stacking

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img_cat = cv2.cvtColor(cv2.imread('resources/cat.png'), cv2.COLOR_BGR2RGB)
img_dog = cv2.cvtColor(cv2.imread('resources/dog.png'), cv2.COLOR_BGR2RGB)
```

```python
# Horizontally stacking two images/arrays of same size
cat_dog_hstacked = np.hstack((img_cat, img_dog))

plt.figure(figsize=(6, 3))
plt.imshow(cat_dog_hstacked)
plt.title('horizontally_stacked')
plt.axis('off')
```

```python
# Vertically stacking two images/arrays of same size
cat_dog_vstacked = np.vstack((img_cat, img_dog))

plt.figure(figsize=(3, 6))
plt.imshow(cat_dog_vstacked)
plt.title('vertically_stacked')
plt.axis('off')
```
