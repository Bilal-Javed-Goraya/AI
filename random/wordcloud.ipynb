```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
text = " Python is a high-level, general-purpose programming\
         language. Its design philosophy emphasizes code\
         readability with the use of significant indentation.\
         Python is used in Machine Learning, Artificial Intelligence\
         Web Dev, Game Development, Scientific computing etc. \
         Python is most widely accepted laguage in Data Science\
         community."
```

```python
wc = WordCloud().generate(text)
plt.axis('off')
plt.imshow(wc)
plt.show()
```
