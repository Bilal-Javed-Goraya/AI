```python
import cv2
```

```python
# Load original image
image = cv2.imread('messi_vs_nld.png')

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Invert the grayscale image to create a negative
negative = 255 - gray

# Blur the negative to smooth it and reduce the amount of detail
blur = cv2.GaussianBlur(negative, (45, 45), sigmaX=0, sigmaY=0)

# Blend the grayscale image with the blurred negative using a color dodge
pencil = cv2.divide(gray, 255-blur, scale=256)
```

```python
# Display util
from PIL import Image
Image.fromarray(pencil)
```
