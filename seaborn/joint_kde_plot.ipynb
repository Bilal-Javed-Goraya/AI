#### ‚ùáÔ∏è Setup

```python
import seaborn as sns
```

#### ‚ùáÔ∏è Dataset

```python
# seaborn provides some builtin datasets. We will use üêß penguins dataset today
# to explore more try this üëâ print(sns.get_dataset_names())
penguins = sns.load_dataset('penguins')
penguins.head(2)
```

#### ‚ùáÔ∏è Joint kde plot

```python
# Let's say we want a joint kde plot of penguin's bill_depth and bill_length
# the semantic we use here is species.
# We are able to see both univariate and bivariate kde plots üëá 
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", kind="kde")
```
