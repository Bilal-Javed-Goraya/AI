#### ‚ùáÔ∏è Setup

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

#### ‚ùáÔ∏è Dataset

```python
# seaborn provides some builtin datasets. We will use geyser dataset today
# to explore more try this üëâ print(sns.get_dataset_names())
geyser = sns.load_dataset('geyser')
geyser.head(2)
```

#### ‚ùáÔ∏è Bivariate kde plot

```python
#Today we see three different ways to visualise the data and how to interpret the results
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Along the contours probability density remains same
sns.kdeplot(ax = axes[0], data=geyser, x="waiting", y="duration", hue="kind")

# Darker the colour, more is the probability density
sns.kdeplot(ax=axes[1], data=geyser, x="waiting", y="duration", hue="kind", fill=True,)

# Brighter the intensity, more is the probability density
sns.kdeplot(ax = axes[2], data=geyser, x="waiting", y="duration", fill=True, thresh=0, levels=100, cmap="mako")
```
