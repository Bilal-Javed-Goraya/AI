#### ‚ùáÔ∏è Dataset 

```python
# seaborn provides some builtin datasets. We will use Tips üíµ dataset today
# to explore more try this üëâ print(sns.get_dataset_names())
tips_df = sns.load_dataset('tips')
tips_df.head(2)
```

#### ‚ùáÔ∏è Violin Plot

```python
# Today, we see how total_bill varies across diffrent days.
# Note: Violin plots use the kernel density estimation of the data points for these plots.
# Check this out üëá 
sns.violinplot(x="day", y="total_bill", data=tips_df)
```

#### ‚ùáÔ∏è Lmplot

```python
# Today, we see how tip varies with on total_bill using a regression plot.
# Notice how behaviour changes for somker and non-smoker.
# Also, we play around with different styling elements.
# Check this out üëá 
g = sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips_df,
               palette=dict(Yes="m", No="g"))
```

#### ‚ùáÔ∏è Box Plot

```python
# Today, we see how total_bill varies across diffrent days in case of men and women.
# Note: the bar in middle of each box represents median and box extends one standard 
# deviation on either side of it.
# Check this out üëá 
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

#### ‚ùáÔ∏è Joint scatter plot

```python
# seaborn provides some builtin datasets. We will use üêß penguins dataset today
# to explore more try this üëâ print(sns.get_dataset_names())
penguins = sns.load_dataset('penguins')
penguins.head(2)
```

```python
# Let's say we want a joint scatter plot of penguin's flipper_length and body_mass
# ideally there must be a direct correlation between the two, check out the plot üëá
sns.jointplot(x='body_mass_g', y='flipper_length_mm', data=penguins, kind='scatter', hue='sex')
```

#### ‚ùáÔ∏è Pair Plot

```python
# Plot pairwise relationships in a dataset.
sns.pairplot(penguins, hue='sex')
```

#### ‚ùáÔ∏è Joint Plot: Hexagonal distribution

```python
# Let's say we want a joint plot of penguin's flipper_length and body_mass
# ideally there must be a direct correlation between the two, check out the plot üëá
sns.jointplot(x='body_mass_g', y='flipper_length_mm', data=penguins, kind='hex')
```
