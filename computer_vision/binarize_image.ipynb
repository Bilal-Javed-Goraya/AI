#### ❇️ Binarise an Image

```python
import cv2
from matplotlib import pyplot as plt
```

```python
# Original RGB image
img_orig = cv2.cvtColor(cv2.imread('lena.png'), cv2.COLOR_BGR2RGB)

# Reading as grayscale image;
img_gray = cv2.imread('lena.png', cv2.IMREAD_GRAYSCALE) 
```

```python
# Apply adaptive thresholding
maxValue_output = 255
blockSize = 99
subtract_from_mean = 10
image_binarized = cv2.adaptiveThreshold(img_gray, 
                                        maxValue_output, 
                                        cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
                                        cv2.THRESH_BINARY, 
                                        blockSize, 
                                        subtract_from_mean)
```

```python
# This util is just used for Display purpose
images = [img_orig, img_gray, image_binarized]
image_names = ['original', 'grayscale', 'binarised']

num_images = len(images) # Maximum number of images to display
num_cols = 3 # Number of columns in display
num_rows = num_images//num_cols # Number of rows in display
plt.figure(figsize=(8,4*num_cols))
for i in range(num_rows*num_cols):
    
    plt.subplot(num_rows, num_cols, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(image_names[i], size=12)
    plt.axis('off') 
```
