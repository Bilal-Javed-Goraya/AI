#### ‚ùáÔ∏è Blurring an Image
Blurring is an important Image augmentation technique used in sevaral computer vision tasks

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Original RGB image
img_original = cv2.cvtColor(cv2.imread('lena.png'), cv2.COLOR_BGR2RGB)
```

```python
# Image blurring is achieved by convolving the image with a low-pass filter kernel. 
# It is useful for removing noise. It actually removes high frequency content 
# (eg: noise, edges) from the image. üëá we have used a 41*41 kernel.
img_blurry = cv2.blur(img_original, ksize=(41,41))
```

```python
# This util is just used for Display purpose
images = [img_original, img_blurry]
image_names = ['img_original', 'img_blurry']

num_images = len(images) # Maximum number of images to display
num_cols = 2 # Number of columns in display
num_rows = num_images//num_cols # Number of rows in display
plt.figure(figsize=(8,4*num_cols))
for i in range(num_rows*num_cols):
    
    plt.subplot(num_rows, num_cols, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(image_names[i], size=12)
    plt.axis('off') 
```
