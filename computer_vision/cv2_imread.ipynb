```python
import cv2
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# by default cv2 reads a 3-channel image in bgr format
img_bgr = cv2.imread('lena.png')

# reading grayscale image; üëá equivalent to cv2.imread('lena.png', 0)
img_gray = cv2.imread('lena.png', cv2.IMREAD_GRAYSCALE) 

# converting BGR to RGB
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
 
# converting BGR to HSV: Hue Saturation Value
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
```

```python
# Display util
images = [img_bgr, img_gray, img_rgb, img_hsv]
image_names = ['img_bgr', 'img_gray', 'img_rgb', 'img_hsv']

num_images = len(images) # Maximum number of images to display
num_cols = 2 # Number of columns in display
num_rows = num_images//num_cols # Number of rows in display

plt.figure(figsize=(6,3*num_cols))
for i in range(num_rows*num_cols):
    
    plt.subplot(num_rows, num_cols, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(image_names[i], size=12)
    plt.axis('off') 
```
